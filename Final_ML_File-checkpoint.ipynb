# Importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import VotingClassifier

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
df.drop('Unnamed: 0' , axis = 1 , inplace = True)
```

# Spliting Independent and Dependent Variables

```python
x = df.drop(columns = 'Churn') # Input features (attributes)
y = df['Churn'] # Target Vector
```

```python
y.head()
```

# Model Training 

# 1.Decision Tree

```python
x_train_dt, x_test_dt, y_train_dt, y_test_dt = train_test_split(x, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
x_train_dt
```

```python
x_test_dt
```

```python
y_train_dt
```

```python
y_test_dt
```

```python
maxdepths = [2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50]
trainAcc = np.zeros(len(maxdepths))
testAcc = np.zeros(len(maxdepths))

index = 0
for depth in maxdepths:
    clf = tree.DecisionTreeClassifier(max_depth=depth)
    clf = clf.fit(x_train_dt, y_train_dt)
    y_predTrain = clf.predict(x_train_dt)
    y_predTest = clf.predict(x_test_dt)
    trainAcc[index] = accuracy_score(y_train_dt, y_predTrain)
    testAcc[index] = accuracy_score(y_test_dt, y_predTest)
    print("Depth = ", depth, ", Training Accuracy = ", trainAcc[index], ", Test Accuracy = ", testAcc[index], 
          ", Diff = ", np.abs((trainAcc[index]-testAcc[index])*100), " %")
    index += 1

    
#########################################
# Plot of training and test accuracies
#########################################
    
plt.plot(maxdepths,trainAcc,'ro-',maxdepths,testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Max depth')
plt.ylabel('Accuracy');
```

```python
# Create a model
dt = DecisionTreeClassifier( criterion='entropy',max_depth = 10, random_state=0)
dt = dt.fit(x_train_dt, y_train_dt)
tree.plot_tree(dt)
```

```python
## Checking Training Accuracy 

print("Training accuracy = ", dt.score(x_train_dt , y_train_dt)*100)
```

```python
## Checking Testing Accuracy 

print("Testing accuracy = ", dt.score(x_test_dt , y_test_dt)*100)
```

# 2.XGBoost Classifier

```python
x_train_xg, x_test_xg, y_train_xg, y_test_xg = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier(objective='binary:logistic')
xg = xg.fit(x_train_xg,y_train_xg)
```

```python
## Checking Training Accuracy 

print("Training accuracy = ", xg.score(x_train_xg , y_train_xg)*100)
```

```python
## Checking Testing Accuracy 

print("Testing accuracy = ", xg.score(x_test_xg , y_test_xg)*100)
```

# 3.Random Forest

```python
x_train_rf, x_test_rf, y_train_rf, y_test_rf = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
maxdepths = [2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50]
trainAcc = np.zeros(len(maxdepths))
testAcc = np.zeros(len(maxdepths))

index = 0
for depth in maxdepths:
    rf = RandomForestClassifier(max_depth=depth)
    rf = rf.fit(x_train_rf, y_train_rf)
    y_predTrain = rf.predict(x_train_rf)
    y_predTest = rf.predict(x_test_rf)
    trainAcc[index] = accuracy_score(y_train_rf, y_predTrain)
    testAcc[index] = accuracy_score(y_test_rf, y_predTest)
    print("Depth = ", depth, ", Training Accuracy = ", trainAcc[index], ", Test Accuracy = ", testAcc[index], 
          ", Diff = ", np.abs((trainAcc[index]-testAcc[index])*100), " %")
    index += 1

    
#########################################
# Plot of training and test accuracies
#########################################
    
plt.plot(maxdepths,trainAcc,'ro-',maxdepths,testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Max depth')
plt.ylabel('Accuracy');
```

```python
from sklearn.ensemble import RandomForestClassifier

model_tree = RandomForestClassifier(n_estimators = 20,max_depth=15)
model_tree.fit(x_train_rf, y_train_rf)
pred = model_tree.predict(x_test_rf)

print("Accuracy Score: ",accuracy_score(pred,y_test_rf))
print("Confusion Matrix:\n",confusion_matrix(pred,y_test_rf))
print("Classification Report: \n",classification_report(pred,y_test_rf))
```

# Multiple ROC-Curves in a single plot

```python
final_roc_auc = []
```

```python
probas1 = dt.predict_proba(x_test_dt)[:, 1]
probas2 = xg.predict_proba(x_test_xg)[:, 1]
probas3 = model_tree.predict_proba(x_test_rf)[:, 1]
probas4 = lr.predict_proba(x_test_lr)[:, 1]
probas5 =gradient_booster.predict_proba(x_test_gbc)[:, 1]
probas6 =naive.predict_proba(x_test_nb)[:, 1]
probas7 =knn.predict_proba(x_test_knn)[:, 1]
```

```python
# get False Positive Rates and True Postive rates for the Classifiers because these will be used to plot the ROC Curve.
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test_dt, probas1)
false_positive_rate2, true_positive_rate2, threshold2 = roc_curve(y_test_xg, probas2)
false_positive_rate3, true_positive_rate3, threshold3 = roc_curve(y_test_rf, probas3)
false_positive_rate4, true_positive_rate4, threshold4 = roc_curve(y_test_lr, probas4)
false_positive_rate5, true_positive_rate5, threshold5 = roc_curve(y_test_gbc, probas5)
false_positive_rate6, true_positive_rate6, threshold6 = roc_curve(y_test_nb, probas6)
false_positive_rate7, true_positive_rate7, threshold7 = roc_curve(y_test_knn, probas7)
```

```python
print('roc_auc_score for DecisionTree: ', roc_auc_score(y_test_dt, probas1))
print('roc_auc_score for XGboost: ', roc_auc_score(y_test_xg, probas2))
print('roc_auc_score for Random Forest Classifier', roc_auc_score(y_test_rf, probas3))
print('roc_auc_score for Logistic regression: ', roc_auc_score(y_test_lr, probas4))
print('roc_auc_score for Gradient Boosting classifier: ', roc_auc_score(y_test_gbc, probas5))
print('roc_auc_score for Naive Bisen Classifier: ', roc_auc_score(y_test_nb, probas6))
print('roc_auc_score for KNN : ', roc_auc_score(y_test_knn, probas7))
```

```python
final_roc_auc.append({'model_name': 'Decision Tree Classifier', 'fpr': false_positive_rate1, 'tpr': true_positive_rate1, 'auc': roc_auc_score(y_test_dt, probas1)})
final_roc_auc.append({'model_name': ' XGboost', 'fpr': false_positive_rate2, 'tpr': true_positive_rate2, 'auc': roc_auc_score(y_test_xg, probas2)})
final_roc_auc.append({'model_name': 'Random Forest Classifier', 'fpr': false_positive_rate3, 'tpr': true_positive_rate3, 'auc': roc_auc_score(y_test_rf, probas3)})
final_roc_auc.append({'model_name': 'Logistic regression', 'fpr': false_positive_rate4, 'tpr': true_positive_rate4, 'auc': roc_auc_score(y_test_lr, probas4)})
final_roc_auc.append({'model_name': 'Gradient Boosting Classifier', 'fpr': false_positive_rate5, 'tpr': true_positive_rate5, 'auc': roc_auc_score(y_test_gbc, probas5)})
final_roc_auc.append({'model_name': 'Naive Bisen Classifier', 'fpr': false_positive_rate6, 'tpr': true_positive_rate6, 'auc': roc_auc_score(y_test_nb, probas6)})
final_roc_auc.append({'model_name': 'KNN', 'fpr': false_positive_rate7, 'tpr': true_positive_rate7, 'auc': roc_auc_score(y_test_knn, probas7)})
```

```python

fig = plt.figure(figsize=(8,6))

for m in final_roc_auc:
    plt.plot(m['fpr'], 
             m['tpr'], 
             label="{}, AUC={:.2f}".format(m['model_name'], m['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("False Positive Rate", fontsize=15)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)

plt.title('ROC Curve Analysis', fontweight='bold', fontsize=15)
plt.legend(prop={'size':10}, loc='lower right')

plt.show()
```
